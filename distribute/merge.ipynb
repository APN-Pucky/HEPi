```python
from hepi.data import data
params = [
    "sleptons.slha",
]
pss = [
    (+1000011, -1000011),  # l_L^+ l_L^-
    (+2000011, -2000011),  # l_R^+ l_R^-
    (+1000015, -1000015),  # l_1^+ l_1^-
]
ms = "mass_1000011_2000011_1000012_1000013_2000013_1000014_1000015_2000015_1000016"
for pa, pb in pss:
    for energy in [13000, 13600]:
        fig, axs = plt.subplots(2, 1, figsize=(6,4), sharex=True, gridspec_kw={'height_ratios': [2, 1]})         # Remove horizontal space between axes         
        fig.subplots_adjust(hspace=0.1)
        name = str(energy) + "_sleptons_" + str(pa) + "_" + str(pb)
        print(name)
        rs_dl_1 = hepi.load_json(open(name + "_NNLL.json"),1)
        rs_dl_2 = hepi.load_json(open(data.get_file(name+ "_NNLL.json")))
        
        rs_dl = pd.concat([rs_dl_1,rs_dl_2],axis=0)
        rs_dl=rs_dl.sort_values(by=[ms]).drop_duplicates()
        
        rs_dl_1_save = rs_dl_1.copy()
        rs_dl_2 = hepi.interpolate_1d(rs_dl_2,ms,"aNNLO_PLUS_NNLL_COMBINED",list(range(150,1000,50)),    only_interpolation=True,
                                                      interpolator="cubic",
                pre=np.log,
                post=np.exp,
                interpolate_lower_uncertainty=False,
                                      )
        rs_dl_1 = hepi.interpolate_1d(rs_dl_1,ms,"aNNLO_PLUS_NNLL_COMBINED",list(range(150,1000,50)),    only_interpolation=True,
                                                      interpolator="cubic",
                pre=np.log,
                post=np.exp,
                interpolate_lower_uncertainty=False,
                                      )
        rs_dl_1_save = hepi.interpolate_1d(rs_dl_1_save,ms,"aNNLO_PLUS_NNLL_NOERR",list(range(150,1000,50)),    only_interpolation=True,
                                                      interpolator="cubic",
                pre=np.log,
                post=np.exp,
                interpolate_lower_uncertainty=False,
                                      )
        hepi.plot(rs_dl_1,ms,"aNNLO_PLUS_NNLL_COMBINED",yscale=1000,show=False,xaxis="",axes=axs[0],tight=False,plot_data=False,fill=True,logy=True,interpolate=True,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="13.0 TeV" if energy == 13000 else "13.6 TeV" )
        hepi.plot(rs_dl_2,ms,"aNNLO_PLUS_NNLL_COMBINED",yscale=1000,show=False,xaxis="",axes=axs[0],tight=False,plot_data=False,fill=True,logy=True,interpolate=True,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="13.0 TeV" if energy == 13000 else "13.6 TeV" )
        hepi.plot(rs_dl_2,ms,"aNNLO_PLUS_NNLL_COMBINED",yscale=1./(rs_dl_1_save["aNNLO_PLUS_NNLL_NOERR"][rs_dl_1_save[ms]>90]), fill=True,show=False,axes=axs[1],tight=False,plot_data=True,logy=False,interpolate=True,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="ratio", yaxis="ratio")
        hepi.write_twiki(rs_dl,ms,"aNNLO_PLUS_NNLL",yscale=1000)
        hepi.write_json(rs_dl[rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull()],hepi.Order.aNNLO_PLUS_NNLL,[ms],name + "_NNLL.json",error_sym=False)
        
        plt.savefig("ratio.pdf")
        plt.show()
```

```python
rs_dl["aNNLO_PLUS_NNLL"] = rs_dl["aNNLO_PLUS_NNLL_NOERR"]
import smpl.plot as plot
def write_twiki(rs_dl, mass, main):
    for i in range(len(rs_dl[main + "_NOERR"])):
        cen = rs_dl[main + "_NOERR"].iloc[i],
        cen = cen[0]
        print( "|" , 
              rs_dl[mass].iloc[i]              , " | " ,
              "{:.4g}".format(cen*1000), " | ",
              "{:.1f}".format((plot.unv(rs_dl[main + "_SCALE"].iloc[i])-cen + plot.usd(rs_dl[main + "_SCALE"].iloc[i]))/cen*100), "|",
              "{:.1f}".format((plot.unv(rs_dl[main + "_PDF"].iloc[i])-cen + plot.usd(rs_dl[main + "_PDF"].iloc[i]))/cen*100), "|",
              
              "{:.1f}".format((plot.unv(rs_dl[main + "_SCALE"].iloc[i])-cen - plot.usd(rs_dl[main + "_SCALE"].iloc[i]))/cen*100), "|",
              "{:.1f}".format((plot.unv(rs_dl[main + "_PDF"].iloc[i])-cen - plot.usd(rs_dl[main + "_PDF"].iloc[i]))/cen*100), "|",
             )
    
write_twiki(rs_dl,ms,"aNNLO_PLUS_NNLL")
```

```python
from hepi.data import data
params = [
    "sleptons.slha",
]
pss = [
    (+1000011, -1000011),  # l_L^+ l_L^-
    (+2000011, -2000011),  # l_R^+ l_R^-
    (+1000015, -1000015),  # l_1^+ l_1^-
]
ms = "mass_1000011_2000011_1000012_1000013_2000013_1000014_1000015_2000015_1000016"
for pa, pb in pss:
    for energy in [13000, 13600]:
        name = str(energy) + "_sleptons_" + str(pa) + "_" + str(pb)
        print(name)
        rs_dl_1 = hepi.load_json(open(name + "_NNLL.json"),1)
        rs_dl_2 = hepi.load_json(open(data.get_file(name+ "_NNLL.json") ))
        
        rs_dl = pd.concat([rs_dl_1,rs_dl_2],axis=0)
        rs_dl=rs_dl.sort_values(by=[ms]).drop_duplicates()
        if energy == 13000:
            fig, axs = plt.subplots(2, 1, figsize=(6,4), sharex=True, gridspec_kw={'height_ratios': [2, 1]})         # Remove horizontal space between axes         
            fig.subplots_adjust(hspace=0.1)
            #hepi.title(li[0],axe=axs[0],cms_energy=False,pdf_info=False,scenario="")
            frac = rs_dl["aNNLO_PLUS_NNLL_NOERR"][rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull()]
        hepi.combined_plot(rs_dl,ms,"aNNLO_PLUS_NNLL",yscale=1000,axes=axs[0],tight=False,show=False,plot_data=False,logy=True,interpolate=True,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="13.0 TeV" if energy == 13000 else "13.6 TeV" )
        hepi.combined_plot(rs_dl,ms,"aNNLO_PLUS_NNLL",yscale=1./frac,axes=axs[1],tight=False,show=False,plot_data=False,logy=False,interpolate=True,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="" )
        axs[0].set_ylim(5e-7*1e3,1*1e3)
        axs[1].set_ylim(0.75,1.5)
        axs[0].set_xlabel(None)
        axs[1].set_xlabel('$m_{\\tilde{l}}$ [GeV]')
        axs[1].set_ylabel('$\\sigma/\\sigma$(13.0 TeV)')
        axs[0].set_ylabel('$\\sigma$ [fb]')
        if energy == 13600:
            plt.savefig(name + ".pdf",bbox_inches ='tight',pad_inches=0)
            plt.show()
                      
```
