# Scenarios

The scenarios are taken from https://twiki.cern.ch/twiki/bin/view/LHCPhysics/SUSYCrossSections.

Modifications:
-   we set the STAUMIX values to maximally mixed for sleptons
-   we manually set the massses of all particles tagged with an @a or {M...} 

```python
#%load_ext nb_black
import hepi

print(hepi.__version__)
import smpl
from smpl import data
import numpy as np
import hepi.resummino as rs
import hepi.util as util
import matplotlib.pyplot as plt
import urllib.request

rs.set_path("~/git/resummino/build")
print(rs.get_path())
rs.default_resummino_runner._check_path()
print(rs.get_version())

test_lo = False
analyse  = True # else generation

plt.rcParams.update({'font.size': 15})
plt.rc('legend', fontsize=12)
#rm tex files
from smpl import io
import os
for _,_,f in io.files(".tex"):
    os.remove(f)
```

```python
params = [
    "sleptons.slha",
]
pss = [
    (+1000011, -1000011),  # l_L^+ l_L^-
    #(+2000011, -2000011),  # l_R^+ l_R^-
    #(+1000015, -1000015),  # l_1^+ l_1^-
]
for pa, pb in pss:
    for energy in [13000, 13600]:
        name = str(energy) + "_sleptons_" + str(pa) + "_" + str(pb)
        hepi.set_output_dir("output_" + name +"/")
        rs.default_resummino_runner.set_output_dir(hepi.get_output_dir())
        for param in params:
            i = hepi.Input(
                hepi.Order.LO if test_lo else hepi.Order.aNNLO_PLUS_NNLL,
                energy,
                pa,
                pb,
                param,
                "PDF4LHC21_40",
                "PDF4LHC21_40",
                1.0,
                1.0,
                id="0.0.0",
                precision=0.001,
                max_iters=50,
            )
            li = [i]
            li = hepi.masses_scan(
                li,
                [
                    1000011,
                    2000011,
                    1000012,
                    1000013,
                    2000013,
                    1000014,
                    1000015,
                    2000015,
                    1000016,
                ],
                range(100, 1010, 100),
            )
            if not test_lo:
                li = hepi.seven_point_scan(li)
                li = hepi.pdf_scan(li)
                # li = hepi.change_where(li, {"precision": 0.0001, "max_iters": 200}, pdfset_nlo=0)
                if not analyse:
                    rs_dl = rs.run(li, skip=False, run=False, parse=False)
                else:
                    rs_dl = rs.run(li, skip=True, run=True, parse=True,ignore_error=True)
                    rs_dl = hepi.pdf_errors(li,rs_dl)
                    rs_dl = hepi.scale_errors(li,rs_dl)
                    rs_dl = hepi.combine_errors(rs_dl)
                    if energy == 13000:
                        fig, axs = plt.subplots(2, 1, figsize=(6,6), sharex=True, gridspec_kw={'height_ratios': [2, 2]})         # Remove horizontal space between axes         
                        fig.subplots_adjust(hspace=0)
                        hepi.title(li[0],axe=axs[0],cms_energy=False,pdf_info=False,scenario="")
                        frac = rs_dl["aNNLO_PLUS_NNLL_NOERR"][rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull()]
                        
                    hepi.mass_plot(rs_dl,"aNNLO_PLUS_NNLL_NOERR",1000011,axes=axs[0],mask = rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull(),tight=False,fmt=" ",next_color=False,show=False,plot_data=False,logy=True,interpolate=True,fill=False,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,interpolate_label="13.0 TeV PDF4LHC21" if energy == 13000 else "13.6 TeV PDF4LHC21" )
                    hepi.mass_plot(rs_dl,"aNNLO_PLUS_NNLL_COMBINED",1000011,axes=axs[0],mask = rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull(),tight=False,fit_fmt=" ",next_color=True,show=False,plot_data=False,logy=True,interpolate=True,fill=True,interpolator='cubic',pre=lambda a  :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="",interpolate_lower_uncertainty=False)
                    hepi.mass_plot(rs_dl,"aNNLO_PLUS_NNLL_NOERR",1000011,axes=axs[1],yscale=1./frac,mask = rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull(),tight=False,yaxis="Ratio",fmt=" ",next_color=False,show=False,plot_data=False,logy=False,interpolate=True,fill=False,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="")
                    hepi.mass_plot(rs_dl,"aNNLO_PLUS_NNLL_COMBINED",1000011,axes=axs[1],yscale=1./frac,mask = rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull(),tight=False,yaxis="Ratio",fit_fmt=" ",next_color=True,show=False,plot_data=False,logy=False,interpolate=True,fill=True,interpolator='cubic',pre=lambda a  :np.log(a.clip(min=1e-22)),post=np.exp,init=False,label="",interpolate_lower_uncertainty=False)
                    
                    if energy == 13000:
                        ypol = rs_dl["aNNLO_PLUS_NNLL_NOERR"][rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull()]
                        xpol = rs_dl["mass_1000011_2000011_1000012_1000013_2000013_1000014_1000015_2000015_1000016"][rs_dl["aNNLO_PLUS_NNLL_COMBINED"].notnull()]
                        polled = smpl.interpolate.interpolate(xpol,ypol,interpolator='cubic',pre=lambda a :np.log(a.clip(min=1e-22)),post=np.exp)
                        
                        dl = hepi.load(urllib.request.urlopen(
                        "https://raw.githubusercontent.com/APN-Pucky/xsec/master/json/pp13_slep_L_NLO%2BNLL.json"
                        ))
                        hepi.plot(dl,"slL","NLO_PLUS_NLL",axes=axs[0],fill=True,logy=True,tight=False,plot_data=False,interpolate=True,interpolator='cubic',pre=lambda a  :np.log(a.clip(min=1e-22)),post=np.exp,interpolate_label="13.0 TeV CT10NLO")
                        hepi.vplot(dl["slL"],dl["NLO_PLUS_NLL"]/polled(dl['slL']),axes=axs[1],yscale=1.,logy=False,fill=True,tight=False,plot_data=False,interpolate=True,interpolator='cubic')
                    
                        dl = hepi.load(urllib.request.urlopen(
                        "https://raw.githubusercontent.com/APN-Pucky/xsec/master/json/pp13_slep_L_NLO%2BNLL_PDF4LHC.json"
                        ))
                        dl = dl[(dl["slL"]<=1000) & (dl["slL"]>=100)]
                        hepi.plot(dl,"slL","NLO_PLUS_NLL_NOERR",axes=axs[0],fill=False,logy=True,tight=False,plot_data=False,interpolate=True,interpolator='cubic',next_color=False,pre=lambda a  :np.log(a.clip(min=1e-22)),post=np.exp,interpolate_label="13.0 TeV PDF4LHC15")
                        hepi.plot(dl,"slL","NLO_PLUS_NLL_COMBINED",axes=axs[0],fill=True,logy=True,tight=False,plot_data=False,interpolate=True,interpolator='cubic',pre=lambda a  :np.log(a.clip(min=1e-22)),post=np.exp,label="")
                        hepi.vplot(dl["slL"].to_numpy(),(dl["NLO_PLUS_NLL_NOERR"]/polled(dl['slL'])).to_numpy(),axes=axs[1],yscale=1.,logy=False,fill=False,tight=False,plot_data=False,next_color=False,interpolate=True,interpolator='cubic')
                        hepi.vplot(dl["slL"].to_numpy(),(dl["NLO_PLUS_NLL_COMBINED"]/polled(dl['slL'])).to_numpy(),axes=axs[1],yscale=1.,logy=False,fill=True,tight=False,plot_data=False,interpolate=True,interpolator='cubic')
                    
                    axs[0].set_ylim(5e-7,1)
                    axs[1].set_ylim(0.6,1.6)
                    axs[0].set_xlabel(None)
                    axs[1].set_xlabel('$m_{\\tilde{l}}$ [GeV]')
                    if energy == 13600:
                        
                        plt.savefig(name + ".pdf",bbox_inches ='tight',pad_inches=0)
                        plt.show()
                    #hepi.write_latex_table_transposed_header(rs_dl,"aNNLO_PLUS_NNLL",param+str(energy)+".tex","mass_1000011")
                    #hepi.write_latex_table_transposed(rs_dl,"aNNLO_PLUS_NNLL",param+str(energy)+".tex",yscale=1000.)
            else:
                rs_dl = rs.run(li, skip=False, run=True, parse=True)
                hepi.mass_plot(rs_dl,"LO",1000011,show=True,interpolator='exp')
```
