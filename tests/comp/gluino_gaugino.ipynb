```python
import hepi
import smpl
import numpy as np
import hepi.madgraph as mg
import matplotlib.pyplot as plt
import hepi.resummino as rs
import time
params = ["mg.in",
          #"param_card_nlo.dat", # Broken Pipe in MG
          #"mastercode_no_gm2.in",
          #"mastercode_with_gm2.in",
          #"VANILLA_param_card.dat"
         ]
model_path="/opt/MG5_aMC_v2_7_0/models/EWKino_NLO_UFO_py3"
#model_path="/opt/MG5_aMC_v2_7_0/models/MSSMatNLO_UFO"
mg.set_path("/opt/MG5_aMC_v2_7_0/")
print (mg.get_path())
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
print (rs.get_path())
```

# LO comparison

```python
ps = [(1000022,1000021),
      (1000023,1000021),
      (1000025,1000021),
      (1000035,1000021),
      (1000024,1000021),
      (-1000037,1000021)]
```

## Scale

```python
for p,pb in ps:
    for param in params:
        i = hepi.Input(hepi.Order.LO,13000,p,pb,param,"CT14lo","CT14lo",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.scan(li,"mu_f",np.around(np.logspace(-1, 1, 4),2))
        li = hepi.scan(li,"mu_r",np.around(np.logspace(-1, 1, 4),2))
        rs_dl = rs.run(li,True,False)
        mg_dl = mg.run(li,False)
        rs_dl['rel_diff_LO'] = (rs_dl["LO"]-mg_dl["LO"])/rs_dl["LO"]
        hepi.scale_plot(rs_dl,["LO"],cont=False)
        hepi.scale_plot(mg_dl,["LO"],cont=True)
        smpl.plot.show()
        hepi.scale_plot(rs_dl,["rel_diff_LO"],cont=False)
        plt.ylim([-0.1,0.1])
        smpl.plot.show()
```

## Mass

```python
for p,pb in ps:
    for param in params:
        i = hepi.Input(hepi.Order.LO,13000,p,pb,param,"CT14lo","CT14lo",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.mass_scan([i],p, np.linspace(1000,4000,16),diff_L_R=None)
        rs_dl = rs.run(li,True,False)
        mg_dl = mg.run(li,False)
        rs_dl['rel_diff_LO'] = (rs_dl["LO"]-mg_dl["LO"])/rs_dl["LO"]

        hepi.mass_plot(rs_dl,"LO",p,logy=True)
        #hepi.mass_plot(rs_dl,p,"lo",logy=True)
        hepi.mass_plot(mg_dl,"LO",p,logy=True)
        smpl.plot.show()
        hepi.mass_plot(rs_dl,"rel_diff_LO",p,logy=False,label="rel_diff_LO",yaxis="rel diff")
        plt.ylim([-0.1,0.1])
        smpl.plot.show()
```
