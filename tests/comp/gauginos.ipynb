```python
import hepi
import smpl
import numpy as np
import hepi.madgraph as mg
import matplotlib.pyplot as plt
import hepi.resummino as rs
import hepi.prospino2 as ps2
import time
params = [
    
          "mastercode_with_gm2.in", # LO OK
          "sps1a1000_mod.in",
          "mg.in",
          "sps1a1000_mod_mod.in",
          #"param_card_nlo.dat", # LO Broken Pipe in MG
          #"mastercode_no_gm2_mod.in", # No Mod Wrong LO, Mod broken pipe
          #"VANILLA_param_card.dat", # MG crash (broken pipe)
          #"scenarioA.in", # crash MG missing mass 23 Z
          #"scenarioB.in", # crash MG missing mass 23 Z
          "scenarioB_mod.in",
         ]
model_path="/opt/MG5_aMC_v2_7_0/models/EWKino_NLO_UFO_py3"
#model_path="/opt/MG5_aMC_v2_7_0/models/MSSMatNLO_UFO"
mg.set_path("/opt/MG5_aMC_v2_7_0/")
print (mg.get_path())
rs.set_path("~/git/resummino_test/")
print (rs.get_path())
ps2.set_path("/home/apn/git/Prospino2/")
print (ps2.get_path())
```

# LO comparison

```python

ps = [ 
      (1000022,1000022),
      (1000022,1000023),
      (1000023,1000023),
      (1000022,1000024),
      (1000022,-1000037),
    
      (1000024,-1000024), # these fail due to stop mixing
      (1000037,-1000037),
      (1000024,-1000037),
      (1000037,-1000024),
    
    
    
    
     ]
```

## Scale

```python
for pa,pb in ps:
    for param in params:
        i = hepi.Input(hepi.Order.LO,13000,pa,pb,param,"cteq6l1","cteq66",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.scan(li,"mu_f",np.around(np.logspace(-1, 1, 4),2))
        li = hepi.scan(li,"mu_r",np.around(np.logspace(-1, 1, 4),2))
        try:
            mg_dl = mg.run(li,True)
            rs_dl = rs.run(li,True,False)
            #mg_dl['LO'] = mg_dl["LO"]*2.
            rs_dl['rel_diff_LO'] = (rs_dl["LO"]-mg_dl["LO"])/rs_dl["LO"]    
        except Exception as e:
            print(e)
            continue
        hepi.scale_plot(rs_dl,["LO"],cont=False)
        hepi.scale_plot(mg_dl,["LO"],cont=True)
        smpl.plot.show()
        hepi.scale_plot(rs_dl,["rel_diff_LO"],cont=False)
        #plt.ylim([-0.1,0.1])
        smpl.plot.show()
```

## Mass

```python
for pa,pb in ps:
    for param in params:
        i = hepi.Input(hepi.Order.LO,13000,pa,pb,param,"cteq6l1","cteq66",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.mass_scan([i],pa, np.linspace(100,1000,14))
        rs_dl = rs.run(li,False,False)
        mg_dl = mg.run(li,False)
        #ps_dl = ps2.run(li,True)
        rs_dl['mg_rel_diff_LO'] = (rs_dl["LO"]-mg_dl["LO"])/rs_dl["LO"]
        #rs_dl['ps_rel_diff_LO'] = (rs_dl["LO"]-ps_dl["LO"])/rs_dl["LO"]

        hepi.mass_plot(rs_dl,"LO",pa,logy=True)
        hepi.mass_plot(mg_dl,"LO",pa,logy=True)
        #hepi.mass_plot(ps_dl,"LO",pa,logy=True)
        smpl.plot.show()
        hepi.mass_plot(rs_dl,"mg_rel_diff_LO",pa,logy=True,label="mg_rel_diff_LO",yaxis="rel diff")
        #hepi.mass_plot(rs_dl,"ps_rel_diff_LO",pa,logy=True,label="ps_rel_diff_LO",yaxis="rel diff")
        #plt.ylim([-0.1,0.1])
        smpl.plot.show()
```

# NLO comparison

```python
params = [
          "mastercode_with_gm2_rot.in",
          "mastercode_with_gm2.in", # LO OK
          "scenarioB_mod.in",
          "sps1a1000_mod.in",
          "sps1a1000_mod_mod.in",
          "mg.in",
          #"param_card_nlo.dat", # LO Broken Pipe in MG
          #"mastercode_no_gm2_mod.in", # No Mod Wrong LO, Mod broken pipe
          #"VANILLA_param_card.dat", # MG crash (broken pipe)
          #"scenarioA.in", # crash MG missing mass 23 Z
          #"scenarioB.in", # crash MG missing mass 23 Z
         ]
ps = [ 
      (1000022,1000022),
      (1000035,1000035),
      (1000035,1000022),
    
      (1000022,1000024),
     ]
```

## Scale

```python

for pa,pb in ps:
    for param in params:
        i = hepi.Input(hepi.Order.NLO,13000,pa,pb,param,"cteq6l1","cteq66",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.scan(li,"mu_f",np.around(np.logspace(-1, 1, 4),2))
        li = hepi.scan(li,"mu_r",np.around(np.logspace(-1, 1, 4),2))
        try:
            skip = True
            mg_dl = mg.run(li,not skip)
            rs_dl = rs.run(li,not skip,False)
            rs_dl['rel_diff_NLO'] = (rs_dl["NLO"]-mg_dl["NLO"])/rs_dl["NLO"]
        except Exception as e:
            print(e)
            continue
            
        hepi.scale_plot(rs_dl,["LO","NLO"],cont=False)
        hepi.scale_plot(mg_dl,["NLO"],cont=True)
        smpl.plot.show()
        hepi.scale_plot(rs_dl,["rel_diff_NLO"],yaxis="rel_diff_NLO",cont=False)
        #plt.ylim([-0.1,0.1])
        smpl.plot.show()
```

## Mass

```python
for pa,pb in ps:
    for param in params:
        print(param)
        i = hepi.Input(hepi.Order.NLO,13000,pa,pb,param,"cteq6l1","cteq66",1., 1.,model_path=model_path)
        li = [i]
        li = hepi.mass_scan([i],pa, np.linspace(100,1000,14))
        try:
            mg_dl = mg.run(li,False)
            rs_dl = rs.run(li,False,False)
            ps_dl = ps2.run(li,False)
            rs_dl['mg_rel_diff_NLO'] = (rs_dl["NLO"]-mg_dl["NLO"])/rs_dl["NLO"]
            rs_dl['ps_rel_diff_NLO'] = (rs_dl["NLO"]-ps_dl["NLO"])/rs_dl["NLO"]
        except Exception as e:
            print(e)
            continue

        hepi.mass_plot(rs_dl,"LO",pa,logy=True,interpolate=True)
        hepi.mass_plot(mg_dl,"LO",pa,logy=True,interpolate=True)
        hepi.mass_plot(ps_dl,"LO",pa,logy=True,interpolate=True)
        hepi.mass_plot(rs_dl,"NLO",pa,logy=True,interpolate=True)
        hepi.mass_plot(mg_dl,"NLO",pa,logy=True,interpolate=True)
        hepi.mass_plot(ps_dl,"NLO",pa,logy=True,interpolate=True)
        smpl.plot.show()
        hepi.mass_plot(rs_dl,"mg_rel_diff_NLO",pa,logy=True,label="mg_rel_diff_NLO",yaxis="rel diff",interpolate=True)
        hepi.mass_plot(rs_dl,"ps_rel_diff_NLO",pa,logy=True,label="ps_rel_diff_NLO",yaxis="rel diff",interpolate=True)
        #plt.ylim([-0.1,0.1])
        smpl.plot.show()
```
