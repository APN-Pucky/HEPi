```python
import hepi
import smpl
import hepi.resummino as rs
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
print (rs.get_path())
```

```python
i = hepi.Input(hepi.Order.NLO,7000,2000002,1000022,"sps1a1000.in","CT14lo","CT14lo",1., 1.,"ppk_UUB")
li = hepi.mass_scan([i],2000002, range(300,1000,50),diff_L_R=20)
dll = rs.run(li)
hepi.mass_vplot(dll,2000002,((dll["p_plus_k"]+dll["lo"])/dll["lo"]),logy=False,label="$K_\\mathrm{p+k}$")
hepi.mass_vplot(dll,2000002,((dll["lo"])/dll["lo"]),logy=False,label="$K_\\mathrm{lo}$")
```
