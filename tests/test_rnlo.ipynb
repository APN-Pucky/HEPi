```python
import hepi
import smpl
import hepi.resummino as rs
rs.set_path("~/git/resummino_rnlo/")
print (rs.get_path())
```

```python
rs.set_path("~/git/resummino_rnlo/")
i = hepi.Input(hepi.Order.NLO,7000,2000002,1000022,"sps1a1000.in","CT14lo","CT14lo",1., 1.,"test_rnlo")
li = hepi.mass_scan([i],2000002, range(300,1000,50),diff_L_R=20)
dll = rs.run(li,True)
hepi.mass_plot(dll,2000002,"lo",yscale=1000,yaxis="$\sigma$ [fb]")
hepi.mass_plot(dll,2000002,"rnlo",yscale=1000,yaxis="$\sigma$ [fb]")
smpl.plot.show()
hepi.mass_vplot(dll,2000002,((dll["rnlo"]+dll["lo"])/dll["lo"]),logy=False,label="$K_\\mathrm{rnlo}$")
hepi.mass_plot(dll,2000002,"K_lo",logy=False,label="$K_\\mathrm{lo}$")
```
