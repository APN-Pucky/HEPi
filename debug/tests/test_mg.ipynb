```python
import hepi
import smpl
import numpy as np
import hepi.madgraph as mg
import hepi.resummino as rs
import matplotlib.pyplot as plt
import time
print (mg.get_path())
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
print (rs.get_path())
```

```python
for p in [2000002,1000002]:
    i = hepi.Input(hepi.Order.NLO,13000,p,1000022,"param_card_nlo.dat","CT14lo","CT14lo",1., 1.,"ISTR1_DR")
    li = [i]
    li = hepi.scan(li,"mu_f",np.around(np.logspace(-1, 1, 4),2))
    li = hepi.scan(li,"mu_r",np.around(np.logspace(-1, 1, 4),2))
    rs_dl = rs.run(li,True,False)
    mg_dl = mg.run(li,False)
    rs_dl['ratio_nlo'] = rs_dl["nlo"]/mg_dl["nlo"]
    hepi.scale_plot(rs_dl,["lo","nlo","rnlo","vnlo","p_plus_k"],cont=False)
    hepi.scale_plot(mg_dl,["nlo"],cont=True)
    smpl.plot.show()
    hepi.scale_plot(rs_dl,["ratio_nlo"],cont=False)
    smpl.plot.show()

    #i = hepi.Input(hepi.Order.NLO,13000,p,1000022,"param_card_nlo.dat","CT14lo","CT14lo",1., 1.,"ISTR1_DR")
    #li = [i]
    #li = hepi.mass_scan([i],p, np.linspace(1000,4000,16),diff_L_R=20)
    #rs_dl = rs.run(li,True,False)
    #mg_dl = mg.run(li,False)
    #rs_dl['ratio_nlo'] = rs_dl["nlo"]/mg_dl["nlo"]
    #hepi.mass_plot(rs_dl,p,"ratio_nlo",logy=False,label="ratio_nlo")
    #smpl.plot.show()
```
