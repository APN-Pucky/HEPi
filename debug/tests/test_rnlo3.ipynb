```python
import hepi
import smpl
import hepi.resummino as rs
import numpy as np
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
print (rs.get_path())
```

```python
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
i = hepi.Input(hepi.Order.NLO,7000,1000002,1000022,"sps1a1000.in","cteq6l1","cteq6l1",1., 1.,"test_rnlo_QQ")
li = hepi.mass_scan([i],1000002, np.linspace(300,950,8),diff_L_R=20)
dll = rs.run(li,True)
hepi.mass_plot(dll,1000002,"lo",yscale=1000,yaxis="$\sigma$ [fb]")
hepi.mass_plot(dll,1000002,"rnlo",yscale=1000,yaxis="$\sigma$ [fb]")
smpl.plot.show()
hepi.mass_vplot(dll,1000002,((dll["rnlo"]+dll["lo"])/dll["lo"]),logy=False,label="$K_\\mathrm{rnlo}$")
hepi.mass_plot(dll,1000002,"K_lo",logy=False,label="$K_\\mathrm{lo}$")
```
