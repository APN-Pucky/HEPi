```python
import hepi
import smpl
import numpy as np
import hepi.madgraph as mg
import matplotlib.pyplot as plt
import hepi.resummino as rs
import time
print (mg.get_path())
rs.set_path("~/git/resummino_ug_to_UX_vNLO/")
print (rs.get_path())
```

```python
for p in [2000002]:
    i = hepi.Input(hepi.Order.NLO,13000,p,1000022,"param_card_nlo.dat","CT14lo","CT14lo",1., 1.,OS="DR")
    li = [i]
    li = hepi.scan(li,"mu_f",np.around(np.logspace(-1, 1, 4),2))
    li = hepi.scan(li,"mu_r",np.around(np.logspace(-1, 1, 4),2))
    rs_dl = rs.run(li,False)
    mg_dll = mg.run(li,False)
    hepi.scale_plot(rs_dl,["lo","nlo"],cont=False)
    hepi.scale_plot(mg_dl,["nlo"],cont=True)
```
